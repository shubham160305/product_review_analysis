```python
# =============================
# FINAL BERT (FAST + STABLE + GPU)
# =============================
import torch
import numpy as np
import pandas as pd
import json
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_curve
from sklearn.preprocessing import label_binarize
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import Dataset, DataLoader
from torch.amp import autocast, GradScaler
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re
# =============================
# GPU SETTINGS
# =============================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.backends.cudnn.benchmark = True
print("Using:", device)

# =============================
# LOAD DATA
# =============================
texts, labels = [], []
with open("amazon_reviews.json", "r", encoding="utf-8") as f:
    for line in f:
        item = json.loads(line)
        text = item.get("summary","") + " " + item.get("reviewText","")
        label = int(item["overall"])
        texts.append(text)
        labels.append(label)

df = pd.DataFrame({"text": texts, "label": labels})

def clean_text(text):
    text = text.lower()
    text = re.sub(r"<.*?>", "", text)          # remove HTML
    text = re.sub(r"http\S+", "", text)        # remove URLs
    text = re.sub(r"[^a-zA-Z0-9.,!? ]", "", text)  # keep useful chars
    text = re.sub(r"\s+", " ", text).strip()   # remove extra spaces
    return text
# =============================
# 3 CLASS
# =============================
def convert(x):
    if x <= 2: return 0
    elif x == 3: return 1
    else: return 2

df["label"] = df["label"].apply(convert)

# balance
balanced = []
for i in range(3):
    balanced.append(df[df.label==i].sample(20000, random_state=42))

df = pd.concat(balanced).sample(frac=1).reset_index(drop=True)

# =============================
# SPLIT
# =============================
train_texts, val_texts, y_train, y_val = train_test_split(
    df.text.tolist(),
    df.label.tolist(),
    test_size=0.2,
    stratify=df.label.tolist(),
    random_state=42
)

# =============================
# TOKENIZER
# =============================
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# =============================
# DATASET (FAST)
# =============================
class ReviewDataset(Dataset):
    def __init__(self, texts, labels):
        self.encodings = tokenizer(
            texts,
            truncation=True,
            padding=True,
            max_length=256
        )
        self.labels = labels
        
    def __len__(self):
        return len(self.labels)
    
    def __getitem__(self, idx):
        return {
            "input_ids": torch.tensor(self.encodings["input_ids"][idx]),
            "attention_mask": torch.tensor(self.encodings["attention_mask"][idx]),
            "labels": torch.tensor(self.labels[idx], dtype=torch.long)
        }

train_loader = DataLoader(
    ReviewDataset(train_texts, y_train),
    batch_size=64,
    shuffle=True,
    num_workers=0,        # 🔥 FIXED (no freeze)
    pin_memory=True
)

val_loader = DataLoader(
    ReviewDataset(val_texts, y_val),
    batch_size=64,
    num_workers=0,
    pin_memory=True
)

# =============================
# MODEL
# =============================
model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=3,
    hidden_dropout_prob=0.3,
    attention_probs_dropout_prob=0.3
).to(device)

# freeze first 8 layers
for name, param in model.bert.named_parameters():
    if "encoder.layer." in name:
        layer_num = int(name.split("encoder.layer.")[1].split(".")[0])
        if layer_num < 8:
            param.requires_grad = False

optimizer = torch.optim.AdamW(
    filter(lambda p: p.requires_grad, model.parameters()),
    lr=2e-5
)

scaler = GradScaler("cuda")

# =============================
# TRACKERS
# =============================
train_losses, val_losses, val_acc_list = [], [], []
best_val_loss = float('inf')
patience, counter = 3, 0

# =============================
# TRAIN
# =============================
epochs = 20

# 🔥 warmup (fix first delay)
model.eval()
dummy = torch.randint(0,100,(1,128)).to(device)
with torch.no_grad():
    model(dummy, attention_mask=torch.ones_like(dummy))

for epoch in range(epochs):
    model.train()
    total_loss = 0
    
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}")
    
    for batch in loop:
        optimizer.zero_grad()
        
        input_ids = batch["input_ids"].to(device, non_blocking=True)
        attention_mask = batch["attention_mask"].to(device, non_blocking=True)
        labels = batch["labels"].to(device, non_blocking=True)
        
        with autocast("cuda"):
            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            loss = outputs.loss
        
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        
        total_loss += loss.item()
        loop.set_postfix(loss=loss.item())

    avg_train_loss = total_loss / len(train_loader)
    train_losses.append(avg_train_loss)

    # =============================
    # VALIDATION
    # =============================
    model.eval()
    val_loss = 0
    preds, true, probs_list = [], [], []
    
    with torch.no_grad():
        for batch in val_loader:
            input_ids = batch["input_ids"].to(device, non_blocking=True)
            attention_mask = batch["attention_mask"].to(device, non_blocking=True)
            labels = batch["labels"].to(device, non_blocking=True)
            
            with autocast("cuda"):
                outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
                val_loss += outputs.loss.item()
            
            logits = outputs.logits
            probs = torch.softmax(logits, dim=1)
            predictions = torch.argmax(logits, dim=1)
            
            preds.extend(predictions.cpu().numpy())
            true.extend(labels.cpu().numpy())
            probs_list.extend(probs.cpu().numpy())
    
    val_loss /= len(val_loader)
    val_losses.append(val_loss)

    val_acc = accuracy_score(true, preds)
    val_acc_list.append(val_acc)

    print(f"Epoch {epoch+1} | Train Loss: {avg_train_loss:.4f} | Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}")

    # early stopping
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        counter = 0
        torch.save(model.state_dict(), "best_model.pt")
    else:
        counter += 1
        if counter >= patience:
            print("⛔ Early stopping triggered")
            break

# =============================
# LOAD BEST MODEL
# =============================
model.load_state_dict(torch.load("best_model.pt"))

# =============================
# RESULTS
# =============================
print("\nClassification Report:\n")
print(classification_report(true, preds, target_names=["Negative","Neutral","Positive"]))

# =============================
# PLOTS
# =============================
os.makedirs("plots", exist_ok=True)

plt.plot(train_losses, label="Train")
plt.plot(val_losses, label="Val")
plt.legend(); plt.title("Loss")
plt.savefig("plots/loss.png"); plt.close()

plt.plot(val_acc_list, label="Val Acc")
plt.legend(); plt.title("Accuracy")
plt.savefig("plots/accuracy.png"); plt.close()

cm = confusion_matrix(true, preds)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.savefig("plots/confusion.png"); plt.close()

true_bin = label_binarize(true, classes=[0,1,2])
probs_np = np.array(probs_list)

plt.figure()
for i in range(3):
    fpr, tpr, _ = roc_curve(true_bin[:,i], probs_np[:,i])
    plt.plot(fpr, tpr)
plt.savefig("plots/roc.png")
plt.close()

print("✅ FAST TRAINING COMPLETED 🚀")
```

```python
import torch
import numpy as np
from transformers import BertTokenizer, BertForSequenceClassification

# =============================
# DEVICE
# =============================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# =============================
# LOAD MODEL
# =============================
model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=3
).to(device)

model.load_state_dict(torch.load("best_model.pt", map_location=device))
model.eval()

# =============================
# TOKENIZER
# =============================
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# =============================
# LABEL MAP
# =============================
labels_map = {
    0: "Negative",
    1: "Neutral",
    2: "Positive"
}

# =============================
# PREDICT FUNCTION
# =============================
def predict(text):
    enc = tokenizer(
        text,
        truncation=True,
        padding=True,
        max_length=256,
        return_tensors="pt"
    ).to(device)

    with torch.no_grad():
        outputs = model(**enc)
        logits = outputs.logits
        probs = torch.softmax(logits, dim=1)

    pred_class = torch.argmax(probs, dim=1).item()
    confidence = probs[0][pred_class].item()

    return {
        "text": text,
        "prediction": labels_map[pred_class],
        "confidence": round(confidence, 4)
    }

# =============================
# TEST
# =============================
if __name__ == "__main__":
    sample = "THE product was perfect"
    
    result = predict(sample)
    print(result)
```

```python
import pandas as pd
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

os.makedirs("plots", exist_ok=True)

# ── LOAD DATA ──────────────────────────────────────────────
texts, labels = [], []
with open("amazon_reviews.json", "r", encoding="utf-8") as f:
    for line in f:
        item = json.loads(line)
        text = item.get("summary", "") + " " + item.get("reviewText", "")
        labels.append(int(item["overall"]))
        texts.append(text)

df = pd.DataFrame({"text": texts, "label": labels})

# ── 3-CLASS CONVERSION ─────────────────────────────────────
df["label"] = df["label"].apply(lambda x: 0 if x <= 2 else (1 if x == 3 else 2))

# ── BALANCE ────────────────────────────────────────────────
target = 20000
df = (
    pd.concat([df[df.label == i].sample(n=target, random_state=42) for i in range(3)])
    .sample(frac=1, random_state=42)
    .reset_index(drop=True)
)

df["text_length"] = df["text"].apply(lambda x: len(x.split()))

# ── LABEL MAP ──────────────────────────────────────────────
label_names = {0: "Negative", 1: "Neutral", 2: "Positive"}
palette     = {0: "#e05252", 1: "#e09a3a", 2: "#5aaa5a"}   # red / orange / green
df["sentiment"] = df["label"].map(label_names)
order = ["Negative", "Neutral", "Positive"]

# ── FIGURE 3 ───────────────────────────────────────────────
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4.5))
fig.subplots_adjust(bottom=0.18, wspace=0.35)

# LEFT — overall histogram
ax1.hist(df["text_length"], bins=50, color="#4a90d9", edgecolor="white", linewidth=0.3)
ax1.set_title("Review Text Length Distribution", fontsize=12, fontweight="bold", pad=10)
ax1.set_xlabel("Number of Words", fontsize=10)
ax1.set_ylabel("Frequency", fontsize=10)
ax1.set_xlim(0, df["text_length"].quantile(0.999))
ax1.tick_params(labelsize=9)
sns.despine(ax=ax1)

# RIGHT — box plot by sentiment class
sns.boxplot(
    data=df,
    x="sentiment",
    y="text_length",
    order=order,
    palette={k: palette[i] for i, k in enumerate(order)},
    width=0.5,
    linewidth=1.2,
    flierprops=dict(marker="o", markersize=3, linestyle="none", markerfacecolor="gray",
                    markeredgecolor="none", alpha=0.5),
    ax=ax2,
)
ax2.set_title("Text Length by Sentiment Class", fontsize=12, fontweight="bold", pad=10)
ax2.set_xlabel("Sentiment Class", fontsize=10)
ax2.set_ylabel("Number of Words", fontsize=10)
ax2.set_ylim(0, df["text_length"].quantile(0.999) * 1.05)
ax2.tick_params(labelsize=9)
sns.despine(ax=ax2)

# SHARED CAPTION
fig.text(
    0.5, 0.04,
    "Figure 3: Text Length Distribution (Overall) and by Sentiment Class (Box Plot)",
    ha="center", fontsize=9, style="italic", color="#333333"
)

plt.savefig("plots/figure3_text_length.png", dpi=180, bbox_inches="tight")
plt.close()
print("✅ Figure 3 saved → plots/figure3_text_length.png")
```
